```python
def best_items(racers):
    """Given a list of racer dictionaries, return a dictionary mapping items to the number
    of times those items were picked up by racers who finished in first place.
    """

    winner_item_counts = {}
    for i in range(len(racers)):

        # The i'th racer dictionary
        racer = racers[i]

        # We're only interested in racers who finished in first
        if racer['finish'] == 1:
            for i in racer['items']:
                # Add one to the count for this item (adding it to the dict if necessary)

                if i not in winner_item_counts:
                    winner_item_counts[i] = 0

                winner_item_counts[i] += 1

        # Data quality issues :/ Print a warning about racers with no name set. We'll take care of it later.
        if racer['name'] is None:
            print("WARNING: Encountered racer with unknown name on iteration {}/{} (racer = {})".format(
                i+1, len(racers), racer['name'])
                 )
    return winner_item_counts

sample = [
    {'name': 'Peach', 'items': ['green shell', 'banana', 'green shell',], 'finish': 3},
    {'name': 'Bowser', 'items': ['green shell',], 'finish': 1},
    {'name': None, 'items': ['mushroom',], 'finish': 2},
    {'name': 'Toad', 'items': ['green shell', 'mushroom'], 'finish': 1},
]

print(best_items(sample))
```

```python
cards1 = ['K', 'A']
cards2 = ['7', '10', 'A']

def hand_total(hand):
    """Helper function to calculate the total points of a blackjack hand.
    """
    total = 0
    # Count the number of aces and deal with how to apply them at the end.
    aces = 0
    for card in hand:
        if card in ['J', 'Q', 'K']:
            total += 10
        elif card == 'A':
            aces += 1
        else:
            # Convert number cards (e.g. '7') to ints
            total += int(card)
    # At this point, total is the sum of this hand's cards *not counting aces*.

    # Add aces, counting them as 1 for now. This is the smallest total we can make from this hand
    total += aces
    # "Upgrade" aces from 1 to 11 as long as it helps us get closer to 21
    # without busting
    while total + 10 <= 21 and aces > 0:
        # Upgrade an ace from 1 to 11
        total += 10
        aces -= 1
    return total

def blackjack_hand_greater_than(hand_1, hand_2):
    total_1 = hand_total(hand_1)
    total_2 = hand_total(hand_2)
    return total_1 <= 21 and (total_1 > total_2 or total_2 > 21)

print("Partita con le carte da gioco <<blackpoint>>")

print("\n")

print("Regole del gioco")
print("Regola 1 - Aggiungere un punto se le carte estratte sono numeri che vannno da 2 a 10.")
print("Regola 2 - Aggiungere dieci punti se le carte estratte sono Assi <<Aces - A>>, Fanti <<Jacks - J>>, Re <<Kings - K>> e Regine <<Queens - Q>>.")
print("Regola 3 - Se il punteggio totale supera il valore di 21 bisogna abbassarlo di un valore di 10 in maniera iterativa inserendo il punteggio residuo relativo al conteggio degli Assi estratti.")

print("\n")

print("Quali carte ha estratto il primo giocatore? {}".format(cards1))
print("Quali carte ha estratto il secondo giocatore? {}".format(cards2))

print("\n")

print("Quanti punti ha ottenuto il primo giocatore? {} punti.".format(hand_total(cards1)))
print("Quanti punti ha ottenuto il secondo giocatore? {} punti.".format(hand_total(cards2)))

print("\n")

print("Il primo giocatore ha vinto contro il secondo giocatore?", blackjack_hand_greater_than(cards1, cards2))
print("Il secondo giocatore ha vinto contro il primo giocatore?", blackjack_hand_greater_than(cards2, cards1))
```

#### Paragrafo 3 $\rightarrow$ **Scegliere tra "loc" e "iloc"** (*Choosing between "loc" and "iloc"*)

Quando **si sceglie** o **si passa** dall'**operatore di etichetta ("loc")** all'**operatore di indice posizionale ("iloc")** e viceversa, c'è un "**inconveniente** ("*gotcha*") che **vale la pena tenere a mente**: i due metodi **utilizzano** schemi di indicizzazione **leggermente diversi**.

L'operatore **"iloc"** utilizza lo schema di indicizzazione **"stdlib"** di Python, nel quale il primo elemento dell'intervallo viene incluso mentre l'ultimo elemento invece viene escluso.

Così allora succederà che l'argomento **"0:10"** applicato all'operatore di indicizzazione **"iloc"** selezionererà la serie di entrate **"0,...,9"**. L'operatore di etichettatura **"loc"**, invece, esegue il processo di indicizzazione in maniera inclusiva. Così allora succederà che l'argomento **"0:10"** applicato all'operatore di etichettatura **"loc"** selezionererà la serie di entrate **"0,...,10"**.

Perché avviene questo cambiamento? Ricordatevi che l'operatore **"loc"** può indicizzare qualsiasi tipologia di **"stdlib"**: le stringhe, per esempio. Se abbiamo un'entrata **"DataFrame"** con valori di indice etichettata come una serie di stringhe del tipo **"Apples, [...,] Potatoes, [...]"** e vogliamo selezionare la voce **"all the alphabetical fruit choices between Apples and Potatoes"**, allora c'è una maniera più conveniente per indicizzare la prima funzione **"df.loc['Apples':'Potatoes']"** invece che òa seconda funzione **"df.loc['Apples', 'Potatoet']"** (ricordandoci che il carattere *t* viene dopo il carattere *s* lungo la serie alfabetica adottata come stringa).

Questa situazione è particolarmente confusionale quando l'indicizzazione dell'entrata **"DataFrame"** è una semplice lista numerica, ad esempio **"$0$, $...$, $1000$"**. In questo caso, la funzione **"df.iloc[0:100]"** restituirà $1000$ entrate, mentre la funzione **"df.loc[0:100]"** ne restituirà invece $1001$! Per fornire $1000$ elementi utilizzando l'operatore **"loc"**, avrete bisogno di accorciare l'intervallo di indicizzazione in una maniera più opportuna cosicchè, in questo caso, richiederete di utilizzare la funzione **"df.loc[0:999]"**.

Per il resto, l'utilizzo di entrambi gli operatori **"loc"** e **"iloc"** si basa sulla loro medesima semantica.

    * **Suggerimento**: Potete utillizare l'operatore **"loc"** oppure l'operatore **"iloc"**. Quando lavorate sulla risposta a questa domanda e a quelle che seguono, tenete presente il seguente **elenco di avvisi ("gotcha")** descritto qui sotto.
    
      * Avviso 1 $\rightarrow$ L'operatore **"iloc"** utilizza la libreria **"stdlib"** del codice *Python* secondo uno **schema di indicizzazione** (*indexing scheme*), dove il primo elemento dell'intervallo viene incluso e l'ultimo escluso. L'operatore **"loc"**, invece indicizza in maniera inclusiva.
      
      * Avviso 2 $\rightarrow$ Questa situazione è particolarmente confusionale quando l'indicizzazione dell'entrata **"DataFrame"** è una semplice lista numerica, ad esempio **"$0$, $...$, $1000$"**. In questo caso, la funzione **"df.iloc[0:100]"** restituirà $1000$ entrate, mentre la funzione **"df.loc[0:100]"** ne restituirà invece $1001$! Per fornire $1000$ elementi utilizzando l'operatore **"loc"**, avrete bisogno di accorciare l'intervallo di indicizzazione in una maniera più opportuna cosicchè, in questo caso, richiederete di utilizzare la funzione **"df.loc[0:999]"**.

* Metodo 2 $\rightarrow$ Il secondo metodo di mappatura definito come **funzione "apply()"**, cioè, come **metodo di applicazione** (*apply method*), è l'esatto equivalente del primo se vogliamo trasformare un' intera entrata **"DataFrame"** chiamando un metodo personalizzato su ogni riga.
  
  * Esempio 71 → Stessa logica dell'Esempio 70 del Tutorial 3 **utilizzando una funzione di applicazione ("apply") al posto della funzione di mappatura ("map") all'interno della funzione predefinita ("def") applicata ad una riga "row" della tabella "reviews" intesa come argomento della funzione prededinita "def" che sfrutta l'algoritmo iterativo del punteggio residuo inteso come la differenza tra il punteggio attuale "row.points" e il punteggio medio "review_points_mean" dichiarato esternamente**.

```python
import numpy as np
import pandas as pd

reviews = pd.read_csv("https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv")

wine_reviews_data_mean_points = reviews.points.mean()
wine_reviews_data_map_points = reviews.points.map(lambda p: p - wine_reviews_data_mean_points)

def remean_points(row):
    row.points = row.points - wine_reviews_data_mean_points
    return row

wine_reviews_data_residual_points = reviews.apply(remean_points, axis = 'columns')

print("\nStampare in maniera riassuntiva il contenuto della tabella del file CSV sui 130 000 prodotti di vino importati nei vari Paesi leggendone la media matematica dei loro punteggi senza il riporto di 0 mediante la funzione <<dataframe_function.mean()>>.\n", wine_reviews_data_mean_points)

print("\nStampare in maniera riassuntiva il contenuto della tabella del file CSV sui 130 000 prodotti di vino importati nei vari Paesi leggendone la mappatura dei loro punteggi con il riporto di 0 mediante la funzione <<dataframe_function.map(lambda p: p - dataframe_function)>>.\n", wine_reviews_data_map_points)

print("\nStampare in maniera riassuntiva il contenuto della tabella del file CSV sui 130 000 prodotti di vino importati nei vari Paesi leggendone il residuo dei loro punteggi senza il riporto di 0 mediante l'algoritmo iterativo <<row.points = row.points - wine_reviews_data_mean_points>> contenuto nella funzione predefinita <<remean_points(row)>>.\n", wine_reviews_data_residual_points)
```

* Esercizio 7 →
  Vorremmo pubblicare queste recensioni di vini sul nostro sito *web*, ma un sistema di valutazione che va da $80$ a $100$ punti è troppo complesso da comprendere: vorremmo tradurle in semplici valutazioni a stelle. Un punteggio di $95$ o superiore vale $3$ stelle, un punteggio di almeno $85$ ma inferiore a $95$ vale 2 stelle. Qualsiasi altro punteggio vale $1$ stella.
  
  Inoltre, la *Canadian Vintners Association* ha acquistato molti annunci sul sito, quindi tutti i vini canadesi dovrebbero ricevere automaticamente $3$ stelle, indipendentemente dai punti.
  
  Creare un'entrata **"Series"** intitolata **"star_ratings"** con il numero di stelle con il numero di stelle corrispondente a ciascuna recensione nel *set* di dati.
  
    * Esempio 81 $\rightarrow$ Stessa logica dell'Esempio 64 del Tutorial 2 **utilizzando una funzione predefinita "def"**.

```python
import numpy as np
import pandas as pd

reviews = pd.read_csv("https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv")

def stars(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

star_ratings = reviews.apply(stars, axis = 'columns')

print("\nStampare in maniera descrittiva il contenuto della tabella del file CSV sui 130 000 prodotti di vino importati nei vari Paesi leggendone la lista di vini che sono classificati da 1 stella a 3 stelle.\n", star_ratings)
```
